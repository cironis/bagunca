### Split input

```python
sudoku_split = sudoku_raw.split("  ")

games_and_solutions = []

for pair in sudoku_split:
    sudoku = {"game": "","solution": ""}
    pair_split = pair.split(' ')
    
    sudoku['game'] = pair_split[0]
    sudoku['solution'] = pair_split[1]
    
    games_and_solutions.append(sudoku)
```

### Print boards

```python
def print_board_from_line(board):
    full_board = ""
    
    for index,char in enumerate(board,1):
            full_board += char
            full_board += " "
            if(((index%27) == 0) & (index < 80)):
                full_board += "\n- - - - - - - - - -\n"
                
            elif ((index%9) == 0):
                full_board += "\n"
            
            elif ((index%3) == 0):
                full_board += "|"
                
    print(full_board)
    

def print_number_board():
    number_board = ""
    for x in range(1,82):
        number_board += "{:02d}".format(x)
        number_board += " "
        if(((x%27) == 0) & (x < 80)):
            number_board += "\n-- -- -- -- -- -- -- -- -- --\n"
        elif ((x%9) == 0):
            number_board += "\n"
        elif ((x%3) == 0):
            number_board += "|"
    print(number_board)
    
def print_board_from_list(board_as_list):
    list_board = ""
    for index, x in enumerate(board_as_list,1):
        list_board += "{:02d}".format(x)
        list_board += " "
        if(((index%27) == 0) & (index < 80)):
            list_board += "\n-- -- -- -- -- -- -- -- -- --\n"
        elif ((index%9) == 0):
            list_board += "\n"
        elif ((index%3) == 0):
            list_board += "|"
    print(list_board)
```

### Create Editable Board

```python
def create_editable_board(board):
    board_as_list = []
    for char in board:
        if char == ".":
            board_as_list.append({})
        else:
            board_as_list.append({int(char)})
    
    board_as_list = set(range(1,82))
    
    row_list = [ [] for _ in range(9) ]
    column_list = [ [] for _ in range(9) ]
    square_list = [ [] for _ in range(9) ]
    
    for index, cell in enumerate(board_as_list,1):
        row_index = 0 if (math.ceil(index/9) == 1) else ((math.ceil(index/9)-1))
        column_index = 8 if ((index%9) == 0) else (index%9)-1

        square_row_index = math.ceil((row_index+1)/3)
        square_column_index = math.ceil((column_index+1)/3)
        square_index = (square_row_index-1)*3+square_column_index -1
        
        #print("index: " + str(index) + " row: " + str(square_row_index) + " column: " + str(square_column_index))
        #print(square_row_index)
        print("index: " + str(index) + " square: " +str(square_index) + " row: " + str(row_index) + " column: " + str(column_index))
```
